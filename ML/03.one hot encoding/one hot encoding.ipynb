```python
import pandas as pd
```

```python
df = pd.read_csv('car info.csv')
```

```python
df
```

```python
# here car model,mileage and age is feature and sell price is label beacuse we want to predict the sell price
# now car Model is a not a neumeric field so we have to convert the car Model field to numeric field.
# converting to neumeric field
dumies = pd.get_dummies(df['car Model'])
```

```python
dumies
#converts the field to numeric field using one hot encoding
```

```python
merge_dataset = pd.concat([df,dumies],axis = 'columns')
```

```python
merge_dataset
# show the mergred data set
```

```python
# now drop the car Model field as it is not necessary
final_table = merge_dataset.drop(['car Model'], axis = 'columns')
```

```python
final_table
```

```python
# label..the value that we want to preict. Taking the feature in y value
y = df['sell price']
```

```python
y
```

```python
# extracting the feature in x
x = final_table.drop(['sell price'],axis = 'columns')
```

```python
x
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# take the linear regression model insatnce
reg = LinearRegression()
```

```python
reg.fit(x,y)
```

```python
# check model accuracy
reg.score(x,y)
```

```python
# show the output for these values
reg.predict([[45000,8,1,0,0,0]])
```
