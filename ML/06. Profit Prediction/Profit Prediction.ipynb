```python
# import
import numpy as np;
import matplotlib.pyplot as plt;
import pandas as pd
```

```python
# importing the data set file
df = pd.read_csv("online.csv")
```

```python
df
```

```python
# data set shape
df.shape
```

```python
#check for null value
df.isnull().sum()
```

```python
# NAN means null..handle nulll (no NAN in this dataset)
# missing : df.Administration.mean()
# df.Administration.fillna(missing)
```

# Seperate x and y for y=mx formula

```python
# extracting the features
x = df.drop(['Profit'],axis=1)
```

```python
x
```

```python
y = df['Profit']
```

```python
y
```

# one hot encoding for the area label

```python
# convert the city column into categorical column
city = pd.get_dummies(x['Area'])
```

```python
city
```

```python
x = x.drop(['Area'],axis=1)
```

```python
x
```

```python
# now concat the one hot encoding value
x = pd.concat([x,city],axis =1)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y, test_size=0.25, random_state = 0)
```

```python
ytrain
```

```python
# take the linear regression model insatnce
#reg = linear_model.LinearRegression()
regressor = LinearRegression()
```

```python
regressor
```

```python
regressor.fit(xtrain,ytrain)
```

```python
# predict the test set result
predict = regressor.predict(xtest)
```

```python
predict
```

# R -Squared Value(accuracy)

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(ytest,predict)
```

```python
score
```
